# Data visualization

```python
# Excluding non-numeric columns
numeric_df = train_data.select_dtypes(include=['number'])

##Calculating Correlation 
correlation=numeric_df.corr()
Positive_Related=correlation.index[abs(correlation["SalePrice"])>0.4]

#Plotting the Correlation in HeatMap for the data columns which has correlation value more than 0.4
plt.figure(figsize=(12,12))
Corr_Heatmap=sns.heatmap(numeric_df[Positive_Related].corr(),annot=True,cmap="GnBu")
```

### Checking outliers

Here checking will be conducted,if there are any outliers in the data columns: OverallQual and GrLiveArea, as they are the data columns that have high correlation value than the rest in the dataset.Box Plot and Scatterplot are most used tools to check if there are outliers in the data column.

```python
plt.figure(figsize=(10,6))
sns.boxplot(x='OverallQual',y='SalePrice',data=train_data)
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x='GrLivArea',y='SalePrice',data=train_data)
```

```python
#Deleting Outliers
train_data=train_data.drop(train_data[(train_data['GrLivArea']>4000)&(train_data['SalePrice']>300000)].index)
```

### Converting textual data to numerical data using LabelEncoder

```python
encoder=preprocessing.LabelEncoder()
def encode_columns(df) :
    colu = df.select_dtypes(include=['object']).columns
    for i in colu:
        df[i]=encoder.fit_transform(df[i])

#Apply to train data
encode_columns(train_data)

#Apply to test data
encode_columns(test_data)
```

```python
train_data.info()
```

```python
test_data.info()
```

### Training Model

```python
model = LinearRegression()
new_data = train_data
X_test = test_data
X_train = new_data.drop("SalePrice", axis=1)
y_train = train_data["SalePrice"]

# Fit the model to the training data
model.fit(X_train, y_train)

# Make predictions on the test set
y_pred = model.predict(X_test)

```

```python
new_y_pred= model.predict(test_data)
```
